## Import modules

```python
import hail as hl
from gnomad_toolbox.load_data import  get_gnomad_release
```

```python
hl.init(backend="local")
```

## Variant data

Supported versions for each data type and reference build are (as of 2025-1-13):

| Genome Build    | Version  | Data Types             |
|-----------------|----------|------------------------|
| GRCh37          | 2.1.1    | exomes, genomes        |
| GRCh38          | 4.1      | exomes, genomes, joint |

For a description of all fields the the HT, see the [Help/FAQ](https://gnomad.broadinstitute.org/help/v4-hts) page.

### v4.1 exomes Hail Table

#### Load the Hail Table


```python
ht = get_gnomad_release(data_type='exomes', version='4.1')
```

#### Print the schema of the Hail Table


```python
ht.describe()
```

#### Show the first 5 variants in the Hail Table


```python
ht.show(5)
```

### v4.1 genomes Hail Table

#### Load the Hail Table


```python
ht = get_gnomad_release(data_type='genomes', version='4.1')
```

#### Print the schema of the Hail Table


```python
ht.describe()
```

#### Show the first 5 variants in the Hail Table

```python
ht.show(5)
```

### v4.1 Joint Frequency Hail Table

The joint frequency Hail table includes frequency for the exomes, genomes, and the exomes+genomes. We have also added statistics for the combined exomes and genomes frequencies, more details on these stats can be found on our [Help](https://gnomad.broadinstitute.org/help/combined-freq-stats) page.

#### Load the Hail Table


```python
ht = get_gnomad_release(data_type='joint', version='4.1')
```

#### Print the schema of the Hail Table


```python
ht.describe()
```

#### Show the first 5 variants in the Hail Table


```python
ht.show(5)
```

## All sites allele numbers

As part of gnomAD v4.1, we [released](https://gnomad.broadinstitute.org/data#v4-all-sites-allele-number) allele number across all callable sites in the gnomAD exomes and genomes. For more information, see our [v4.1 blog post](https://gnomad.broadinstitute.org/news/2024-04-gnomad-v4-1/#allele-numbers-across-all-possible-sites).

### Exomes all sites allele number Hail Table

#### Load the Hail Table


```python
ht = get_gnomad_release(data_type='exomes', version='4.1', dataset="all_sites_an")
```

#### Print the schema of the Hail Table


```python
ht.describe()
```

#### Show the first 5 variants in the Hail Table


```python
ht.show(5)
```

### Genomes all sites allele number Hail Table

#### Load the Hail Table


```python
ht = get_gnomad_release(data_type='genomes', version='4.1', dataset="all_sites_an")
```

#### Print the schema of the Hail Table


```python
ht.describe()
```

#### Show the first 5 variants in the Hail Table


```python
ht.show(5)
```
